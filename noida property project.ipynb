```python
def remove_outlier_sqft(df):
    df_output = pd.DataFrame()
    for key,subdf in data.groupby('address'):
        m = np.mean(subdf.price_per_square)
    
        st = np.std(subdf.price_per_square)
    
        gen_df = subdf[(subdf.price_per_square > (m-st)) &(subdf.price_per_square<=(m+st))]
        df_output = pd.concat([df_output,gen_df],ignore_index= True)
    return df_output
data = remove_outlier_sqft(data)  
```

```python
def bhk_outlier_remover(df):
    exclude_indices = np.array([])
    for location, location_df in data.groupby('address'):
        bhk_stats ={}
        for bhk, bhk_df in location_df.groupby('bedroom'):
            bhk_stats[bhk]={
                'mean':np.mean(bhk_df.price_per_square),
                'std': np.std(bhk_df.price_per_square),
                'count': bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby('bedroom'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices,bhk_df[bhk_df.price_per_square<(stats['mean'])].index.values)
                return df.drop(exclude_indices,axis ='index')
```

```python
data = bhk_outlier_remover(data)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import accuracy_score,precision_score, recall_score,f1_score, r2_score
```

```python
# sparse=True (default in older versions):
# Returns a SciPy sparse matrix — efficient in memory if most entries are zeros (which is common in one-hot encoding).
# sparse=False:
# Returns a dense NumPy array — all values are stored explicitly in memory, even zeros.
```

```python
#column Transfromer 
# catgorical value to numberical value 
column_trans = make_column_transformer(
    (OneHotEncoder(sparse_output=False, handle_unknown='ignore'), ['address']),
    remainder='passthrough')


# Why handle_unknown='ignore' IMPORTANT 🚨
# Production / Streamlit app me jab new address aayega jo training me nahi tha, tab error aata hai:
# Found unknown categories during transform
```

```python
scaler = StandardScaler()
```
